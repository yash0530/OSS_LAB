# Operations

There are lots of operations with pandas that will be really useful to you, but don't fall into any distinct category. Let's show them here in this topic:


**DataFrame.head(n=5)**

Return the first n rows.

This function returns the first n rows for the object based on position. It is useful for quickly testing if your object has the right type of data in it.

Parameters:	
n : int, default 5
Number of rows to select.

Returns:	
obj_head : same type as caller
The first n rows of the caller object.

```python
df = pd.DataFrame({'col1':[1,2,3,4],'col2':[444,555,666,444],'col3':['abc','def','ghi','xyz']})
df.head()
```

### Info on Unique Values

```python
df['col2'].unique()      #gives unique non repeated values
```

```python
df['col2'].nunique()      #gives number of unique values 
```

**Index.value_counts(normalize=False, sort=True, ascending=False, bins=None, dropna=True)**

The resulting object will be in descending order so that the first element is the most frequently-occurring element. Excludes NA values by default.

Parameters:	
normalize : boolean, default False
If True then the object returned will contain the relative frequencies of the unique values.

sort : boolean, default True
Sort by values.

ascending : boolean, default False
Sort in ascending order.

bins : integer, optional
Rather than count values, group them into half-open bins, a convenience for pd.cut, only works with numeric data.

dropna : boolean, default True
Donâ€™t include counts of NaN.

```python
df['col2'].value_counts()
```

### Selecting Data

```python
#Select from DataFrame using criteria from multiple columns
newdf = df[(df['col1']>2) & (df['col2']==444)]
```
